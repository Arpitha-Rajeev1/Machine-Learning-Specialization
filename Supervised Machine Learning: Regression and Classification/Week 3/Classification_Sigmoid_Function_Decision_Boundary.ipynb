```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_common import plot_data
```

```python
x_train = np.array([0, 1, 2, 3, 4, 5]) # single feature
y_train = np.array([0, 0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3,0.5], [2,2], [1,2.5]]) # 2 features
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
print(f'x_train with single feature: {x_train}')
print(f'y_train with single feature: {y_train}')
print(f'X_train2 with two features: {X_train2}')
print(f'y_train2 with two features: {y_train2}')
```

```python
pos = y_train == 1
neg = y_train == 0
print(pos)
print(neg)
```

```python
fig,ax = plt.subplots(1,3,figsize=(12,4))
ax[0].scatter(x_train[pos], y_train[pos], marker='x',s=80,c='red',label="y=1")
ax[0].scatter(x_train[neg], y_train[neg], marker='o',s=80,c='blue',label="y=0")
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('one variable plot')
ax[0].legend()

ax[1].scatter(x_train,y_train) # we should print true and false separately because this results in second figure where 
# both y = 1 and y = 0 have the same color

ax[2].scatter(X_train2[y_train2==1,0], X_train2[y_train2==1,1], marker='x', s=80, color='red',label='y=1')
ax[2].scatter(X_train2[y_train2==0,0], X_train2[y_train2==0,1], marker='o', s=80, color='blue',label='y=0')
ax[2].set_xlabel('x', fontsize=12)
ax[2].set_title('two variable plot')
ax[2].legend()
plt.show()
```

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
```

```python
z_tmp = np.arange(-10, 11)
y = sigmoid(z_tmp)
np.set_printoptions(precision=3) 
print("Input (z), Output (sigmoid(z))")
print(np.c_[z_tmp, y])
```

```python
plt.plot(z_tmp, y, c="b")
plt.title("Sigmoid function")
plt.ylabel('sigmoid(z)')
plt.xlabel('z')
```

```python
# Decision Boundary for the line xo + x1 - 3 = 0
# Choose values between 0 and 6
x0 = np.arange(0,6)

x1 = 3 - x0
fig,ax = plt.subplots(1,1,figsize=(5,4))
# Plot the decision boundary
ax.plot(x0,x1, c="b")
ax.axis([0, 4, 0, 3.5])

# Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

# Plot the original data
plot_data(X_train2,y_train2,ax)
ax.set_ylabel(r'$x_1$')
ax.set_title('Decision Boundary')
ax.set_xlabel(r'$x_0$')
plt.show()
```
